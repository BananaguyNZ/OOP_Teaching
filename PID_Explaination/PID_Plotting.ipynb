```python
from Simulated_Motor import simulate_motor
import plotly.graph_objects as go
```

```python
def plot_result(time, position, speed, power, error, target_position=100, title="PID Result"):
    # Position graph
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=time,
        y=position,
        mode="lines",
        name="Motor Position"
    ))

    fig.add_trace(go.Scatter(
        x=time,
        y=[target_position] * len(time),
        mode="lines",
        name="Target Position",
        line=dict(dash="dash")
    ))

    fig.update_layout(
        title=title,
        xaxis_title="Time (seconds)",
        yaxis_title="Position"
    )

    fig.show()

    # Error graph
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=time,
        y=error,
        mode="lines",
        name="Error"
    ))

    fig.update_layout(
        title="Error Over Time",
        xaxis_title="Time (seconds)",
        yaxis_title="Error"
    )

    fig.show()

    # Power graph
    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=time,
        y=power,
        mode="lines",
        name="Motor Power"
    ))

    fig.update_layout(
        title="Motor Power Over Time",
        xaxis_title="Time (seconds)",
        yaxis_title="Power"
    )

    fig.show()
```

```python
time, position, speed, power, error = simulate_motor(
    Kp=5,
    Ki=0,
    Kd=0,
    total_time = 30
)

plot_result(
    time,
    position,
    speed,
    power,
    error,
    title="P Control Only: Kp=2"
)
```

```python
settings = [
    [0.1, 0, 0, "P only, no I"],
    [0.1, 0, 2, "P only, strong I"],
    [0.1, 0, 5, "P only, very weak I"],
]

fig = go.Figure()

for Kp, Ki, Kd, label in settings:
    time, position, speed, power, error = simulate_motor(Kp, Ki, Kd, total_time=100)

    fig.add_trace(go.Scatter(
        x=time,
        y=position,
        mode="lines",
        name=label
    ))

fig.add_trace(go.Scatter(
    x=time,
    y=[100] * len(time),
    mode="lines",
    name="Target",
    line=dict(dash="dash")
))

fig.update_layout(
    title="PID Tuning Comparison",
    xaxis_title="Time (seconds)",
    yaxis_title="Position"
)

fig.show()
```
